```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Loading lendingclub dataset which is ready to train and test i.e cleaned and encoded (Numberical)
df = pd.read_csv("/content/encodedVersion_dataset (1).csv")
df
```

```python
#Spliting into Features and telling which should be our Target
X = df.drop('loan_status', axis=1)
y = df['loan_status']  # Must be 0 and 1 (binary)
```

```python
#helps in balancing the datas not to get biased decision
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Remapping for before splitting into train/test to make XG boost model for better recognising or understanding to run model
y = y.replace({1: 0, 2: 1})
```

```python
# Stratified split ensures balance in train/test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# building up the neural network by layers of 2
#model = Sequential()
#model.add(Dense(32, input_dim=X.shape[1], activation='relu'))  # input + 1st hidden
#model.add(Dense(16, activation='relu'))                        # 2nd hidden layer
#model.add(Dense(1, activation='sigmoid'))                      # Output layer
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
#setting up the classification method, optimizing algorithm and its expected metrics in what values (Accuracy)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training process starts here by initialising EPOCHs and setting batch size
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
#Evaluating the trained model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n Test Accuracy: {accuracy:.4f}")
```

```python
#Loading lendingclub dataset which is ready to train and test i.e cleaned and encoded (Numberical)
df = pd.read_csv("/content/Encoded_prosperDataset (1).csv")
df
```

```python
#Spliting into Features and telling which should be our Target
X = df.drop('LoanStatus', axis=1)
y = df['LoanStatus']  # Must be 0 and 1 (binary)
```

```python
#helps in balancing the datas not to get biased decision
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Remapping for before splitting into train/test to make XG boost model for better recognising or understanding to run model
y = y.replace({1: 0, 2: 1})
```

```python
# Stratified split ensures balance in train/test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# building up the neural network by layers of 2
model = Sequential()
model.add(Dense(32, input_dim=X.shape[1], activation='relu'))  # input + 1st hidden
model.add(Dense(16, activation='relu'))                        # 2nd hidden layer
model.add(Dense(1, activation='sigmoid'))                      # Output layer
```

```python
#setting up the classification method, optimizing algorithm and its expected metrics in what values (Accuracy)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training process starts here by initialising EPOCHs and setting batch sizemodel.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)
```

```python
#Evaluating the trained model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\n Test Accuracy: {accuracy:.4f}")
```
